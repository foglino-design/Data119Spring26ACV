```python
import pandas as pd
```

```python
cards = pd.read_csv('data/cards_merged.csv')
```

```python
print(cards.isnull().sum())
cards.loc[cards['maxEvolutionLevel'].isnull(), 'maxEvolutionLevel'] = 0
cards.describe()
```

```python
cards.head()
```

```python
cards.info()
# replacing null values in order to convert datatypes
cards = cards.fillna(-999)
```

```python
# converting datatype of numeric columns with no non-integer values
cards.info()
```

```python
print(cards['rarity'].unique())
print(cards['type'].unique())
print(cards['mobility'].unique())
print(cards['targets'].unique())
print(cards['attack_type'].unique())
print(cards['rarity'].value_counts())
print(cards['type'].value_counts())
print(cards['mobility'].value_counts())
print(cards['targets'].value_counts())
print(cards['attack_type'].value_counts())
```

```python
cards.loc[cards['attack_type'].isnull()]
```

```python
card_dum = cards.copy().drop(index = [104, 119])
card_dum['groupCard'] = card_dum['groupCard'].replace('True', 1).replace('False', 0)
card_dum['attack_type'] = card_dum['attack_type'].replace('splash', 1).replace('single', 0)
card_dum['rarity'] = card_dum['rarity'].replace('common', 0).replace('rare', 1).replace('epic', 2).replace('legendary', 3).replace('champioin', 4)
card_dum = cards.astype({'maxEvolutionLevel': 'int',
                         'elixirCost': 'int',
                         'damage11': 'int',
                         'hitpoints': 'int',
                         'groupCard': 'int',
                         'attack_type': 'int',
                         'rarity': 'int',
                         'groupCard': 'int'
                        })
```

```python
card_dum = pd.get_dummies(card_dum, columns = ['type', 'mobility', 'targets'], drop_first = True, dtype = int)
card_dum.head()
```
